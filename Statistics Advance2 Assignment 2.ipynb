## Question 1.Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`. After generating the list, find the following:
##   (i) Write a Python function to calculate the mean of a given list of numbers.Create a function to find the median of a list of numbers
##   (ii) Develop a program to compute the mode of a list of integers
##   (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
##   (iv) Write a Python function to find the geometric mean of a list of positive numbers.
##   (v) Create a program to calculate the harmonic mean of a list of values.
##   (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
##  (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.

````python
import numpy as np
import statistics
from scipy import stats

# Generate a list of 100 integers between 90 and 130
int_list = np.random.randint(90, 131, 100)

# (i) Calculate Mean
def calculate_mean(num_list):
    return sum(num_list) / len(num_list)

mean_value = calculate_mean(int_list)
print(f"Mean: {mean_value}")

# (i) Calculate Median
def calculate_median(num_list):
    num_list.sort()
    n = len(num_list)
    if n % 2 == 0:
        median = (num_list[n//2 - 1] + num_list[n//2]) / 2
    else:
        median = num_list[n//2]
    return median

median_value = calculate_median(int_list)
print(f"Median: {median_value}")

# (ii) Calculate Mode
def calculate_mode(num_list):
    return statistics.mode(num_list)

mode_value = calculate_mode(int_list)
print(f"Mode: {mode_value}")

# (iii) Calculate Weighted Mean
weights = np.random.rand(100)
def calculate_weighted_mean(num_list, weight_list):
    return np.average(num_list, weights=weight_list)

weighted_mean_value = calculate_weighted_mean(int_list, weights)
print(f"Weighted Mean: {weighted_mean_value}")

# (iv) Calculate Geometric Mean
def calculate_geometric_mean(num_list):
    return stats.gmean(num_list)

geometric_mean_value = calculate_geometric_mean(int_list)
print(f"Geometric Mean: {geometric_mean_value}")

# (v) Calculate Harmonic Mean
def calculate_harmonic_mean(num_list):
    return stats.hmean(num_list)

harmonic_mean_value = calculate_harmonic_mean(int_list)
print(f"Harmonic Mean: {harmonic_mean_value}")

# (vi) Calculate Midrange
def calculate_midrange(num_list):
    return (min(num_list) + max(num_list)) / 2

midrange_value = calculate_midrange(int_list)
print(f"Midrange: {midrange_value}")

# (vii) Calculate Trimmed Mean
def calculate_trimmed_mean(num_list, trim_percentage):
    num_list.sort()
    trim_count = int(trim_percentage * len(num_list) / 100)
    trimmed_list = num_list[trim_count:-trim_count]
    return np.mean(trimmed_list)

trimmed_mean_value = calculate_trimmed_mean(int_list, 10)
print(f"Trimmed Mean (10%): {trimmed_mean_value}")
````

## Question 3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

````python
import numpy as np

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initialize a discrete random variable.

        Parameters:
        values (list): List of possible values.
        probabilities (list): List of corresponding probabilities.
        """
        self.values = np.array(values)
        self.probabilities = np.array(probabilities)

        # Check if probabilities sum to 1
        assert np.isclose(np.sum(self.probabilities), 1), "Probabilities must sum to 1"

    def expected_value(self):
        """
        Calculate the expected value of the discrete random variable.

        Returns:
        float: Expected value.
        """
        return np.sum(self.values * self.probabilities)

    def variance(self):
        """
        Calculate the variance of the discrete random variable.

        Returns:
        float: Variance.
        """
        expected_val = self.expected_value()
        return np.sum((self.values - expected_val) ** 2 * self.probabilities)

    def standard_deviation(self):
        """
        Calculate the standard deviation of the discrete random variable.

        Returns:
        float: Standard deviation.
        """
        return np.sqrt(self.variance())

# Example usage
if __name__ == "__main__":
    # Define a discrete random variable with values and probabilities
    values = [1, 2, 3, 4, 5]
    probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

    # Create an instance of the DiscreteRandomVariable class
    rv = DiscreteRandomVariable(values, probabilities)

    # Calculate expected value, variance, and standard deviation
    expected_val = rv.expected_value()
    variance = rv.variance()
    std_dev = rv.standard_deviation()

    print(f"Expected Value: {expected_val}")
    print(f"Variance: {variance}")
    print(f"Standard Deviation: {std_dev}")
````

## Question 4.Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

````python
import numpy as np
import random

# Define the possible outcomes and their probabilities
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Function to roll a die
def roll_die():
    return random.choice(outcomes)

# Simulate rolling the die n times
def simulate_rolls(n):
    return [roll_die() for _ in range(n)]

# Calculate expected value
def calculate_expected_value(outcomes, probabilities):
    return np.sum(np.array(outcomes) * np.array(probabilities))

# Calculate variance
def calculate_variance(outcomes, probabilities):
    expected_val = calculate_expected_value(outcomes, probabilities)
    return np.sum((np.array(outcomes) - expected_val) ** 2 * np.array(probabilities))

# Number of simulations
n = 100000

# Simulate rolls
rolls = simulate_rolls(n)

# Calculate empirical expected value and variance
empirical_expected_value = np.mean(rolls)
empirical_variance = np.var(rolls)

# Calculate theoretical expected value and variance
theoretical_expected_value = calculate_expected_value(outcomes, probabilities)
theoretical_variance = calculate_variance(outcomes, probabilities)

print(f"Empirical Expected Value: {empirical_expected_value}")
print(f"Empirical Variance: {empirical_variance}")
print(f"Theoretical Expected Value: {theoretical_expected_value}")
print(f"Theoretical Variance: {theoretical_variance}")
````

## Question 5.Create a Python function to generate random samples from a given probability distribution (e.g.,binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np
import scipy.stats as stats

def generate_samples(dist, params, size):
    """
    Generate random samples from a given probability distribution.

    Parameters:
    dist (str): Distribution name (binomial, poisson, normal, exponential).
    params (list): Distribution parameters.
    size (int): Sample size.

    Returns:
    samples (numpy array): Random samples.
    mean (float): Sample mean.
    variance (float): Sample variance.
    """
    if dist == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, size)
    elif dist == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, size)
    elif dist == 'normal':
        mu, sigma = params
        samples = np.random.normal(mu, sigma, size)
    elif dist == 'exponential':
        lam = params[0]
        samples = np.random.exponential(lam, size)
    else:
        raise ValueError("Invalid distribution name.")

    mean = np.mean(samples)
    variance = np.var(samples)

    return samples, mean, variance

# Example usage
dist_name = 'binomial'
params = [10, 0.5]  # n, p
sample_size = 1000

samples, mean, variance = generate_samples(dist_name, params, sample_size)

print(f"Distribution: {dist_name}")
print(f"Parameters: {params}")
print(f"Sample Size: {sample_size}")
print(f"Sample Mean: {mean}")
print(f"Sample Variance: {variance}")

# Example usage
dist_name = 'poisson'
params = [10, 0.5]  # lam
sample_size = 1000

samples, mean, variance = generate_samples(dist_name, params, sample_size)

print(f"Distribution: {dist_name}")
print(f"Parameters: {params}")
print(f"Sample Size: {sample_size}")
print(f"Sample Mean: {mean}")
print(f"Sample Variance: {variance}")
````

## Question 8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    x (float or numpy array): Point(s) at which to evaluate the PDF.
    mean (float): Mean of the normal distribution.
    std_dev (float): Standard deviation of the normal distribution.

    Returns:
    float or numpy array: PDF value(s) at the given point(s).
    """
    return norm.pdf(x, loc=mean, scale=std_dev)

# Example usage
mean = 0
std_dev = 1

x = np.linspace(-3, 3, 100)
pdf_values = calculate_normal_pdf(x, mean, std_dev)

# Plot the PDF
plt.plot(x, pdf_values)
plt.title("Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()

# Calculate PDF at specific points
x_values = [-1, 0, 1]
for x in x_values:
    pdf = calculate_normal_pdf(x, mean, std_dev)
    print(f"PDF at x = {x}: {pdf}")
````

## Question 9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

def calculate_exponential_cdf(x, lambda_param):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.

    Parameters:
    x (float or numpy array): Point(s) at which to evaluate the CDF.
    lambda_param (float): Rate parameter of the exponential distribution.

    Returns:
    float or numpy array: CDF value(s) at the given point(s).
    """
    return 1 - np.exp(-lambda_param * x)

# Example usage
lambda_param = 1  # rate parameter

x = np.linspace(0, 5, 100)
cdf_values = calculate_exponential_cdf(x, lambda_param)

# Plot the CDF
plt.plot(x, cdf_values)
plt.title("Exponential Distribution CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.show()

# Calculate CDF at specific points
x_values = [1, 2, 3]
for x in x_values:
    cdf = calculate_exponential_cdf(x, lambda_param)
    print(f"CDF at x = {x}: {cdf}")

# Alternative using SciPy's expon.cdf
x = np.linspace(0, 5, 100)
cdf_values_scipy = expon.cdf(x, scale=1/lambda_param)

# Plot the CDF using SciPy
plt.plot(x, cdf_values_scipy)
plt.title("Exponential Distribution CDF using SciPy")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.show()
````

## Question 10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
import math

def calculate_poisson_pmf(k, lambda_param):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution.

    Parameters:
    k (int or numpy array): Number of occurrences.
    lambda_param (float): Expected value (rate parameter) of the Poisson distribution.

    Returns:
    float or numpy array: PMF value(s) at the given number of occurrences.
    """
    return np.exp(-lambda_param) * (lambda_param ** k) / math.factorial(k)

# Example usage
lambda_param = 5  # expected value

k_values = np.arange(0, 15)
pmf_values = [calculate_poisson_pmf(k, lambda_param) for k in k_values]

# Plot the PMF
plt.bar(k_values, pmf_values)
plt.title("Poisson Distribution PMF")
plt.xlabel("Number of Occurrences")
plt.ylabel("Probability")
plt.show()

# Calculate PMF at specific points
k_values = [2, 5, 8]
for k in k_values:
    pmf = calculate_poisson_pmf(k, lambda_param)
    print(f"PMF at k = {k}: {pmf}")

# Alternative using SciPy's poisson.pmf
k_values = np.arange(0, 15)
pmf_values_scipy = poisson.pmf(k_values, lambda_param)

# Plot the PMF using SciPy
plt.bar(k_values, pmf_values_scipy)
plt.title("Poisson Distribution PMF using SciPy")
plt.xlabel("Number of Occurrences")
plt.ylabel("Probability")
plt.show()
````

## Question 11. M A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.
## To generate the data use the following command:
## ```python
## import numpy as np
## # 50 purchases out of 1000 visitors
## old_layout = np.array([1] * 50 + [0] * 950)
## # 70 purchases out of 1000 visitors  
## new_layout = np.array([1] * 70 + [0] * 930)
  ```
# Apply z-test to find which layout is successful.

````python
import numpy as np
from scipy.stats import norm

# Generate data
np.random.seed(0)  # For reproducibility
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate proportions (conversion rates)
old_proportion = np.mean(old_layout)
new_proportion = np.mean(new_layout)

# Calculate pooled proportion
pooled_proportion = (np.sum(old_layout) + np.sum(new_layout)) / (len(old_layout) + len(new_layout))

# Calculate standard error
standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1/len(old_layout) + 1/len(new_layout)))

# Calculate z-score
z_score = (new_proportion - old_proportion) / standard_error

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Print results
print(f"Old Layout Conversion Rate: {old_proportion:.2%}")
print(f"New Layout Conversion Rate: {new_proportion:.2%}")
print(f"Pooled Proportion: {pooled_proportion:.2%}")
print(f"Standard Error: {standard_error:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject null hypothesis: New layout has a significantly higher conversion rate.")
else:
    print("Fail to reject null hypothesis: No significant difference in conversion rates.")
````

## Question 12. A tutoring service claims that its program improves students' exam scores. A sample of students who participated in the program was taken, and their scores before and after the program were recorded.
## Use the below code to generate samples of respective arrays of marks:
## ```python
## before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
## after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])
## ```
## Use z-test to find if the claims made by tutor are true or false.

````python
import numpy as np
from scipy.stats import norm

# Generate samples
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate means
before_mean = np.mean(before_program)
after_mean = np.mean(after_program)

# Calculate standard deviations
before_std = np.std(before_program, ddof=1)
after_std = np.std(after_program, ddof=1)

# Calculate pooled standard deviation
pooled_std = np.sqrt(((len(before_program) - 1) * before_std**2 + (len(after_program) - 1) * after_std**2) / (len(before_program) + len(after_program) - 2))

# Calculate standard error
standard_error = pooled_std * np.sqrt(1/len(before_program) + 1/len(after_program))

# Calculate z-score
z_score = (after_mean - before_mean) / standard_error

# Calculate p-value (one-tailed test)
p_value = 1 - norm.cdf(z_score)

# Print results
print(f"Before Program Mean: {before_mean:.2f}")
print(f"After Program Mean: {after_mean:.2f}")
print(f"Pooled Standard Deviation: {pooled_std:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject null hypothesis: Tutoring service improves students' exam scores.")
else:
    print("Fail to reject null hypothesis: No significant improvement in exam scores.")
````

## Question 15. A company is testing two different website layouts to see which one leads to higher click-through rates. Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.
## Use the following data:
## ```python
## layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
## layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

````python
import numpy as np
from scipy import stats

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    # Calculate means
    layout_a_mean = np.mean(layout_a_clicks)
    layout_b_mean = np.mean(layout_b_clicks)

    # Calculate standard deviations
    layout_a_std = np.std(layout_a_clicks, ddof=1)
    layout_b_std = np.std(layout_b_clicks, ddof=1)

    # Calculate pooled standard deviation
    pooled_std = np.sqrt(((len(layout_a_clicks) - 1) * layout_a_std**2 + (len(layout_b_clicks) - 1) * layout_b_std**2) / (len(layout_a_clicks) + len(layout_b_clicks) - 2))

    # Calculate standard error
    standard_error = pooled_std * np.sqrt(1/len(layout_a_clicks) + 1/len(layout_b_clicks))

    # Calculate t-statistic
    t_statistic = (layout_b_mean - layout_a_mean) / standard_error

    # Calculate degrees of freedom
    degrees_of_freedom = len(layout_a_clicks) + len(layout_b_clicks) - 2

    # Calculate p-value (one-tailed test)
    p_value = 1 - stats.t.cdf(t_statistic, degrees_of_freedom)

    return t_statistic, degrees_of_freedom, p_value

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

t_statistic, degrees_of_freedom, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject null hypothesis: Layout B leads to higher click-through rates.")
else:
    print("Fail to reject null hypothesis: No significant difference in click-through rates.")
````

## Question 18. An HR department wants to investigate if there's a gender-based salary gap within the company. Develop a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically significant difference between the average salaries of male and female employees.
## Use the below code to generate synthetic data:
## ```python
## # Generate synthetic salary data for male and female employees
## np.random.seed(0)  # For reproducibility
## male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
## female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
def analyze_salary_gap(male_salaries, female_salaries):
    # Calculate means
    male_mean = np.mean(male_salaries)
    female_mean = np.mean(female_salaries)

    # Calculate standard deviations
    male_std = np.std(male_salaries, ddof=1)
    female_std = np.std(female_salaries, ddof=1)

    # Calculate pooled standard deviation
    pooled_std = np.sqrt(((len(male_salaries) - 1) * male_std**2 + (len(female_salaries) - 1) * female_std**2) / (len(male_salaries) + len(female_salaries) - 2))

    # Calculate standard error
    standard_error = pooled_std * np.sqrt(1/len(male_salaries) + 1/len(female_salaries))

    # Calculate t-statistic
    t_statistic = (female_mean - male_mean) / standard_error

    # Calculate degrees of freedom
    degrees_of_freedom = len(male_salaries) + len(female_salaries) - 2

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), degrees_of_freedom))

    return t_statistic, degrees_of_freedom, p_value

t_statistic, degrees_of_freedom, p_value = analyze_salary_gap(male_salaries, female_salaries)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject null hypothesis: Statistically significant salary gap exists.")
else:
    print("Fail to reject null hypothesis: No statistically significant salary gap.")

plt.hist(male_salaries, alpha=0.5, label='Male')
plt.hist(female_salaries, alpha=0.5, label='Female')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.legend()
plt.show()
````
